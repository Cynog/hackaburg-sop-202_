# Basic FFT in Python

```python
# import required packages
from scipy.fft import fft, fftfreq
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
import os
from dotenv import load_dotenv

# load path to data folder
load_dotenv()
DATA = os.getenv("DATA")
```

## Example FFT on overlapping sine waves

```python
# Number of sample points
samplerate = 48000
t_end = 1
n_points = samplerate * t_end

# generate the signal data
t = np.linspace(0, t_end, samplerate*t_end)
v = 2.0*np.sin(40 * 2*np.pi * t) + 3.0*np.sin(2400 * 2*np.pi * t) + 0.5*np.sin(4000 * 2*np.pi * t)

# write wav test file
wavfile.write('test.wav', samplerate, v)

# plot the signal data
plt.plot(t, v)
plt.show()

# apply fft
yf = fft(v)[:n_points//2]
xf = fftfreq(n_points, 1/samplerate)[:n_points//2]

# plot the according fft
plt.plot(xf, 2.0/n_points * np.abs(yf))
plt.grid()
plt.show()
```

```python
# load in a sample wav file
samplerate, v = wavfile.read(os.path.join(DATA, "samples/test.wav"))
n_points = len(v)
t_end = n_points / samplerate
print("samplerate: ", samplerate)
print("t_end: ", t_end)

# plot the signal
t = np.linspace(0, t_end, n_points)
plt.plot(t, v)
plt.show()

# apply fft
yf = fft(v)[:n_points//2]
xf = fftfreq(n_points, 1/samplerate)[:n_points//2]

# plot the according fft
plt.plot(xf[:1000], 2.0/n_points * np.abs(yf[:1000]))
plt.grid()
plt.show()
```
